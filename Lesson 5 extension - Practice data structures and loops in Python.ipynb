We will create a data structure to hold our Germplasm data (I have updated it to be a little bit more complex... now a germplasm may hold TWO alleles - i.e. one germplasm has connections to more than one gene)

Represent these data in Python - create a **single variable** that contains all of this data (using lists and dictionaries).

Create loops over that data structure that can print the three tables to the screen.

  gene  |  gene_name  |  embl
 --- | ---  | --- 
  DEF  | Deficiens  | https://www.ebi.ac.uk/ena/data/view/AB516402
  AP3  | Apetala3   |   https://www.ebi.ac.uk/ena/data/view/AF056541
  AG  | Agamous   |   https://www.ebi.ac.uk/ena/data/view/AL161549
  

```python
# put your code here... take it one step at a time!
# do just one table (an easy one, like the gene table)

DEF = {'gene': 'DEF', 'name': 'Deficiens', 'embl': 'http://AB123434'}
AP3 = {'gene': 'AP3', 'name': 'Apetala3' , 'embl': 'http://AF056541'}
AG  = {'gene': 'AG',  'name': 'Agamous'  , 'embl': 'http://AL161549'}
 
gene_table = [DEF, AP3, AG]

print("gene\tgene_name\tembl")

for gene in gene_table:
    #print(gene['gene'], "\t", gene['name'], "\t", gene['embl'])
    #print("%s\t%s\t%s" % (gene['gene'], gene['name'], gene['embl']))
    print("{0}\t{1}\t{2}".format(gene['gene'], gene['name'], gene['embl']))
    

Stock1 = {'germplasm_id': 2, 'amount': 5, 'date': '10/5/2013', 'location': 'Room 2234'}
Stock2 = {'germplasm_id': 1, 'amount': 9.8, 'date': '12/1/2015', 'location': 'Room 998'}

stocks = [Stock1, Stock2]

print("")
print("gerplasm_id\tamount\tdate\tlocation")
for stock in stocks:
    print("%d\t%f\t%s\t%s" % (stock['germplasm_id'], stock['amount'], stock['date'], stock['location']))



##id 	taxonid 	alleles 	genes
#1 	4150 	def-1 	DEF
#2 	3701 	ap3, ag 	AP3, AG    
Germ1 = {'id': 1, 'taxon': 4150, 'alleles': ['def-1'], 'genes': ['DEF']}
Germ2 = {'id': 2, 'taxon': 3701, 'alleles': ['ap3', "ag"], 'genes': ['AP3', "AG"]}

germs = [Germ1, Germ2]

print("")
print("")
for germ in germs:
    alleles = germ['alleles']
    genes = germ['genes']
    allelestring = ""
    for allele in alleles:
        allelestring += allele + ","
    genestring = ""
    for gene in genes:
        genestring += gene + ","
    print("%d\t%d\t%s\t%s" % (germ['id'], germ['taxon'], allelestring, genestring))

# BETTER SOLUTION    
Germ1 = {'id': 1, 'taxon': 4150, 'alleles': 'def-1', 'genes': DEF}
Germ2 = {'id': 2, 'taxon': 3701, 'alleles': 'ap3', 'genes': AP3}
Germ3 = {'id': 2, 'taxon': 3701, 'alleles': 'ag', 'genes': AG}

print(Germ1['genes']['embl'] + " IT IS HERE")

print("")
print("")
germs = [Germ1, Germ2, Germ3]

for thisgerm in germs:
    website = thisgerm['genes']['embl']
    print("the website for the gene of allele %s is %s" % (thisgerm['alleles'], website))
```
